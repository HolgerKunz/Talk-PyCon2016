### <h3 style="text-align: center;">Let's define a functions to read all dicom files from a given directory.<br />this function reads all files ends to *.dcm and returns a list of strings (their path)</h3>

```python
def read_files(path, extension = '*.dcm'):
    """ 
    Read files with specific extension in a dirctory, 
    """
    if path[-1] != '/':
        path += '/'
    directory = path + extension
    # finds all files with specific format, 
    # here, we are a bit more interested in dicom files
    files = glob.glob(directory)        
    files.sort()

    return files
```

<h1>More Objective,</h1>
<p>&nbsp;</p>
<p>GenArray class, generates a 3D array of CT scan files. We might want to specify a smaller part of those images, so, we consider some limations in X, Y, Z directions.&nbsp;</p>
<h2 style="text-align: center;">Not scary at all, it's only a Loop :)</h2>

```python
class GenArray:
    def __init__(self, directory, 
                 start = 1, ends = 100, 
                 xrng = (0,512) , yrng = (0,512)):
        self.directory = directory
        self.start = start
        self.ends = ends
        self.xrng = xrng 
        self.yrng = yrng 
        
        self.gen()
        
    def gen(self):
        """
        reads files from given directory, then returns a 3D array. 
        This array is somehow like a HU(x, y, z) function. every spacial
        coordinate has a specific HU value, so we can access to it's 
        grayscale, and apply image processing functions and so forth. 
        """
        #pre-allocating
        self.arr = np.zeros((self.xrng[1] - self.xrng[0],  # x - dimension
                             self.yrng[1] - self.yrng[0],  # y - dimension
                             self.ends - self.start))      # z - dimension
        
        cnt = 0
        for i in range(self.start,self.ends):
            # a temporary variable
            tmp_dcm = dicom.read_file(self.directory[i])
            # assigning values to pre-allocated array
            self.arr[:,:,cnt] = tmp_dcm.pixel_array[self.xrng[0]:self.xrng[1],
                                                    self.yrng[0]:self.yrng[1]]
            cnt += 1 
    
    def size(self):
        return self.arr.shape
```

<h2 style="color: #2e6c80;">How to Recognize Bones?</h2>
<p>A very simple approach is<span style="color: #ff0000;"><strong> thresholding filters</strong></span>. Thresholding filters return an array of bool type varibles; it seimply<span style="color: #00ff00;"><em><strong> filters a range of data</strong></em></span> and returns <span style="background-color: #00ccff;"><em>True</em></span> for them, and anything except that range of data will be <em><span style="background-color: #00ccff;">False</span></em>. If $I(x,y)$ is an image, and $T_0$ is a constant, then $B(x,y)$ is a binary array that returns True, only for values greater that $T_0$</p>

<h3 style="text-align: center;">
$
B(x,y) = 
\begin{cases}   
True  &  I(x,y) \ge T_0 \\
False  &  I(x,y) < T_0 \\
\end{cases}
$
</h3>

<h2 style="color: #2e6c80;">Other forms of thresholding:</h2>
<p>The boundaries and limits are opt to you! You can design complex filters, return True for a particular value and False, otherwise. Look at another example: </p>

<h3 style="text-align: center;">
$
B(x,y) = 
\begin{cases}   
True  &  I(x,y) \in [T_1, T_2] \cup [T_3, T_4] \\
False  &  otherwise \\
\end{cases}
$
</h3>

<h1 style="color: #5e9ca0;">Canny Filter</h1>
<h2 style="color: #2e6c80;">What is canny Filter?</h2>
<p>The Canny edge detector is an edge detection operator that uses a multi-stage algorithm to detect a wide range of edges in images.<sup>1</sup></p>
<h2 style="color: #2e6c80;">Gaussian Filter:</h2>
<p>Since all edge detection results are easily affected by image noise, it is essential to filter out the noise to prevent false detection caused by noise. To smooth the image, a Gaussian filter is applied to convolve with the image.</p>
<p>a sample gaussian filter:</p>
<h3 style="text-align: center;">$\mathbf{B} = \frac{1}{159} \begin{bmatrix} 2 & 4 & 5 & 4 & 2 \\4 & 9 & 12 & 9 & 4 \\5 & 12 & 15 & 12 & 5 \\4 & 9 & 12 & 9 & 4 \\2 & 4 & 5 & 4 & 2\end{bmatrix} * \mathbf{A}$</h3>
<h2 style="color: #2e6c80;">Finding the intensity gradient of the image:</h2>
<p>An edge in an image may point in a variety of directions, so the Canny algorithm uses four filters to detect horizontal, vertical and diagonal edges in the blurred image. The edge detection operator (Roberts, Prewitt, Sobel for example) returns a value for the first derivative in the horizontal direction ($Gx$) and the vertical direction ($Gy$). From this the edge gradient and direction can be determined</p>
<h3 style="text-align: center;">$\mathbf{G} = \sqrt{ {\mathbf{G}_x}^2 + {\mathbf{G}_y}^2 }$</h3>
<h3 style="text-align: center;">$\mathbf{\Theta} = \operatorname{atan2}\left(\mathbf{G}_y, \mathbf{G}_x\right)$</h3>
<p>&nbsp;</p>

<h1 style="color: #5e9ca0;">Binary Image Processing</h1>
<h2 style="color: #2e6c80; text-align: left;">Segmentation:</h2>
<p><strong>Erode:</strong> An eroding filter removes outer parts of an interconnected region. To apply this filter, we use a kernel like $B$:</p>
<h3 style="text-align: center;">$I \oplus_b B$</h3>
<p>&nbsp;</p>
<p><strong>Dilate:</strong> same as eoding, but this filter expands the region according to the kernel $B$:</p>
<h3 style="text-align: center;">$I \ominus_b B$</h3>
<p>&nbsp;</p>
<p><strong>Fill holes:</strong> after finding edges, we wish to fill inside the detected region. As a result, we have a binary mask represents bone.</p>

### <h2 style="color: #2e6c80; text-align: left;">Import from Scikit-Image and Scipy</h2>

```python
from skimage.morphology import binary_dilation, binary_erosion
from skimage.morphology import square

from scipy.ndimage.morphology import binary_fill_holes
```

```python
img = femur.arr[:,:,1]

img_canny = canny(img, sigma=2, low_threshold=200)
img_dilate = binary_dilation(img_canny, square(3))
img_erode = binary_erosion(img_dilate, square(3))
img_fill = binary_fill_holes(img_erode)

plt.figure(figsize=(14,8))

plt.subplot(231)
plt.imshow(img, plt.cm.gray)
plt.title('Original Image')

plt.subplot(232)
plt.imshow(img_canny, plt.cm.gray, interpolation='nearest')
plt.title('after applying canny')

plt.subplot(233)
plt.imshow(img_dilate, plt.cm.gray, interpolation='nearest')
plt.title('dilated edges')

plt.subplot(234)
plt.imshow(img_erode, plt.cm.gray, interpolation='nearest')
plt.title('eroded edges')

plt.subplot(235)
plt.imshow(img_fill, plt.cm.gray, interpolation='nearest')
plt.title('fill holes')

plt.subplot(236)
plt.imshow(img*img_fill, plt.cm.gray, interpolation='nearest')
plt.title('remove surrounding')
```

<h1 style="color: #5e9ca0;">Be Classy</h1>
<h2 style="color: #2e6c80; text-align: left;">$Segment$ class does the same procedure to all sections</h2>
<p>Also, it's a simple loop, inhertis some bahvours from $GenArray$ class</p>

```python
class Segment(GenArray):
    def __init__(self, GenArrayObj,
                 sigma = 2, low_threshold=200):
        
        self.sigma = sigma
        self.low_threshold = low_threshold
        self.arr = GenArrayObj.arr
        self.size = GenArrayObj.size()
        self.gen_mask()
        
    def seg_sect(self, img):
        img_canny = canny(img, sigma=self.sigma,
                          low_threshold=self.low_threshold)
        
        img_dilate = binary_dilation(img_canny, square(3))
        img_erode = binary_erosion(img_dilate, square(3))
        img_fill = binary_fill_holes(img_erode)
        
        return img_fill
    
    def gen_mask(self):
        slices = self.size[-1]
        
        # pre-allocating mask
        self.mask = np.zeros_like(self.arr, dtype=np.bool)
        
        for i in range(slices):
            self.mask[:, :, i] = self.seg_sect(self.arr[:, :, i])
```

<h1 style="color: #5e9ca0;">Biomechanics</h1>
<h2 style="color: #2e6c80; text-align: left;">Finite Elements Method (FEM)</h2>
<p>Using finite elements, gives us the power to perform simulations over complex, and non-homogen geometries like bone. FEM, converts a continious problem to a discrete problem</p>
<h2 style="color: #2e6c80; text-align: left;">Meshes</h2>
<p>A FEM problem, works with small and simple geometries called mesh. these meshes together, represent the above mentioned geometry. We assume these meshes are cubic like voxels in CT scan image. In a hetrogenous material, every mesh has a different mechanical properties.</p>
<h2 style="color: #2e6c80; text-align: left;">Mechanical Properties</h2>
<p>We assume, mechanical properties is a function of HU unit. <br />we calculate density of K2HPO from calibration phantom<br />then calculate density of ash using the following formula:</p>

<h3 style="text-align: center;">
$
\begin{align}
\rho_{ash} = 1.22 \rho_{KHP}  + 0.0526
\end{align}
$
</h3>

<p>and then, calculate mechanical properties of bone</p>

<h3 style="text-align: center;">$E(\rho_{ash}) = 
\begin{cases}   
33900 \rho_{ash}^{2.20}  &  \rho_{ash} \leq 0.27 \\
10200 \rho_{ash}^{2.01}  &  \rho_{ash} \geq 0.6 \\
5037  \rho_{ash} & otherwise
\end{cases}$</h3>

### <h2 style="color: #2e6c80; text-align: left;">Let's Write The Functions</h2> <p>Looks even more COOL with Lambda</p>

```python
# calculate slope and interception after calibration
# nevermind, not in our scope
phantom_slope = 1
phantom_intercept = -1023

# functions : 

rho_KHP = lambda HU: phantom_slope * HU + phantom_intercept
rho_ash = lambda rho_KHP : 1.22 * rho_KHP / 1000. + 0.0526

# Elastic modulus    
E_modul = lambda r: (np.logical_and(r>0, r<=0.27) * 33900*r)**2.20 + \
            ((r>=0.60) * 10200*r)**2.01 + \
            (np.logical_and(r>0.27, r<0.60) * (5307*r + 469)) + (r<0) * 15000
```

<h2 style="color: #2e6c80; text-align: left;">Distribution of $E$ and $\rho$</h2>


```python
rho_ = rho_ash(rho_KHP(img))
E = E_modul(rho_)

plt.figure(figsize=(15, 5))

plt.subplot(1,2,1)
plt.imshow(rho_*img_fill)
plt.colorbar()
plt.title('rho')

plt.subplot(1,2,2)
plt.imshow(E*img_fill)
plt.colorbar()
plt.title('E')
```

```python
from skimage.segmentation import find_boundaries as fbn

from copy import deepcopy


def bin_boundary(arr, method='surf'): 
    """
        calculate a new set of points are located in B array 
        this set only containts boundary points (voxels) 
        and inner voxels are omited 
        using BB instead of B can decrease cost of other functions and 
        algorithms in this notebook 
        
        Input : 
            arr:  a 3D binary array 
            method : 'surf' generate a closed surface (default) and 
                     'cont' genr8s contours 
                     
        Output : 
            BB: a 3*n array contain location of each point [x,y,z]
    """
    B = deepcopy(arr)
    if method == 'surf': 
        for lyr in range(B.shape[0]-1):
            B[lyr] = np.logical_or(np.logical_and(np.logical_xor(B[lyr],B[lyr+1])
                                                ,B[lyr])
                                 ,fbn(B[lyr]))
        return B
    
    elif method == 'cont':
        for lyr in range(B.shape[0]-1):
            B[lyr] = fbn(B[lyr])
        return B
```
